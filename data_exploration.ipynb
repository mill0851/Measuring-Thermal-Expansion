```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import openpyxl
```

```python
PATH = 'modern_interferometry_data_incomplete.xlsx'
workbook = openpyxl.load_workbook(PATH, data_only=True)
```

## Loading in Aluminum Data

```python
TRIAL_COLUMNS = [('A','B', 1), ('C','D', 2), ('E','F', 3)]
ROW_LIMITS = (4, 60)

aluminum = pd.DataFrame()
aluminum_data = workbook['Aluminum']

for trial in TRIAL_COLUMNS:
    temp = []
    counts = []

    for row in range(ROW_LIMITS[0], ROW_LIMITS[1]):
        temp_val = aluminum_data[trial[0] + str(row)].value
        count_val = aluminum_data[trial[1] + str(row)].value
        temp.append(temp_val)
        counts.append(count_val)

    aluminum['Temp ' + str(trial[2])] = temp
    aluminum['Counts ' + str(trial[2])] = counts

aluminum.head()

```

## Loading in Alumina Data

```python
TRIAL_COLUMNS = [('A','B', 1), ('C','D', 2), ('E','F', 3)]
ROW_LIMITS = (4, 60)

alumina = pd.DataFrame()
alumina_data = workbook['Alumina']

for trial in TRIAL_COLUMNS:
    temp = []
    counts = []

    for row in range(ROW_LIMITS[0], ROW_LIMITS[1]):
        temp_val = alumina_data[trial[0] + str(row)].value
        count_val = alumina_data[trial[1] + str(row)].value
        temp.append(temp_val)
        counts.append(count_val)

    alumina['Temp ' + str(trial[2])] = temp
    alumina['Counts ' + str(trial[2])] = counts

alumina.head()
```

## Plotting All 3 Trials of Both Sets of Data

```python
plt.figure(figsize=(30,20))
plt.subplot(2,3,1)
plt.scatter(aluminum['Temp 1'], aluminum['Counts 1'])
plt.title('Aluminum Trial 1')
plt.xlabel('Temperature (C)')
plt.ylabel('Counts')
plt.grid()

plt.subplot(2,3,2)
plt.scatter(aluminum['Temp 2'], aluminum['Counts 2'])
plt.title('Aluminum Trial 2')
plt.xlabel('Temperature (C)')
plt.ylabel('Counts')
plt.grid()

plt.subplot(2,3,3)
plt.scatter(aluminum['Temp 3'], aluminum['Counts 3'])
plt.title('Aluminum Trial 3')
plt.xlabel('Temperature (C)')
plt.ylabel('Counts')
plt.grid()

plt.subplot(2,3,4)
plt.scatter(alumina['Temp 1'], alumina['Counts 1'])
plt.title('Alumina Trial 1', size=20)
plt.xlabel('Temperature (C)')
plt.ylabel('Counts')
plt.grid()

plt.subplot(2,3,5)
plt.scatter(alumina['Temp 2'], alumina['Counts 2'])
plt.title('Alumina Trial 2')
plt.xlabel('Temperature (C)')
plt.ylabel('Counts')
plt.grid()

plt.subplot(2,3,6)
plt.scatter(alumina['Temp 3'], alumina['Counts 3'])
plt.title('Alumina Trial 3')
plt.xlabel('Temperature (C)')
plt.ylabel('Counts')
plt.grid()

plt.show()
```
